```python
import pandas as pd
import numpy as np
```

```python
alpha = 2.0
results_path = f"../outputs/run_infojigsaw/alpha={alpha}/max_depth=inf_tolerance=0.001/seed=1234/results.csv"
results = pd.read_csv(results_path)
results = results[results["turn"] > 1]
results.groupby(["model","sample_id"]).apply(lambda x: x.loc[x["turn"].idxmax(),"nll"]).reset_index().groupby("model").agg(["mean", "std"])
```
